# Matplotlib
## Powerful Python library that allows to data visualization.

### Import libraries numpy as np and matplotlib as plt

```python
import numpy as np
import matplotlib.pyplot as plt
```

<h2>

</h2>
We create the variables that contain the values we wish to manage.

x variable will contain the values from -2*pi to 2*pi

sin_y will refer to sin of x and cos_y will refer cos of x

fig, ax will create the figure and the axis

<h2>
    <b>
        Then:
    </b>
</h2>

 ax.plot var will refer to all the variables to create the sinÂ´s graphic, with the label "sin" on blue color

 Next one will do the same thing but with cos instead of sin and on red color.

 ax.set_title will display a title

 ax.legend() is a function to display the legend

 ax.grid(True) shows the square box

 plt.show() displays all together

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

sin_y = np.sin(x)
cos_y = np.cos(x)

fig, ax = plt.subplots()

ax.plot(x, sin_y, label='Sin', color='blue')

ax.plot(x, cos_y, label='Cos', color='red')

ax.set_title('Sin and cos functions')

ax.legend()

ax.grid(True)

plt.show()
```
